## Visualização dos Dados

### Vendas America do Norte por ano

```python
df.groupby("Ano")["Vendas América do Norte"].sum().sort_values(ascending=True).plot.barh(title="Vendas America do Norte por ano")
plt.xlabel("Vendas")
plt.ylabel("Ano");
```

### Vendas 2008

```python
vendas2008 = df[df["Ano"] == 2008]
```

```python
vendas2008.head()
```

### Vendas no Japão por ano

```python
df.groupby("Ano")["Vendas Japão"].sum().sort_values(ascending=True).plot.barh(title="Vendas no Japão por ano")
plt.xlabel("Vendas")
plt.ylabel("Ano");
```

### Jogos com mais de 10 milhões de vendas

```python
games_10_sales = df[df["Vendas Globais"] > 10].sort_values(by = "Vendas Globais", ascending = False).head(15)

fig = px.bar(games_10_sales, x="Ano", y="Vendas Globais", color="Nome", title="Jogos com mais de 10 milhões de vendas", width=1300, height=500)
fig.show()
```

### Desenvolvimento de venda ao longo dos anos

```python
y = df["Vendas Globais"]
x = df["Ano"]
plt.figure(figsize=(20,10))
plt.bar(x , y)
plt.ylabel('Vendas globais em milhões')
plt.grid()
plt.title('Desenvolvimento de vendas ao longo dos anos');
```

### Vendas historicas por editoras

```python
vendas_df = df[["Ano","Vendas Globais","Editora"]].sort_values(by = "Vendas Globais", ascending= False).head(100)

plt.figure(figsize=(20,10))
sns.lineplot(data = vendas_df, x = vendas_df["Ano"], y = vendas_df["Vendas Globais"], hue="Editora")
plt.ylabel('Vendas globais em milhões')
plt.title("Vendas globais históricas por editor");
```

### Vendas Globais por Editor

```python
topEditoras = df[df["Editora"].str.contains('Nintendo')==False]
topEditoras = topEditoras[topEditoras["Vendas Globais"] >= 4]
topEditoras = topEditoras[["Editora", "Vendas Globais"]]
topEditoras.groupby("Editora").sum().plot.bar()
plt.ylabel("Vendas globais em milhões")
plt.title("Vendas globais por editor");
```

### As industrias de jogos em declínio

```python
plt.title("As indústrias de jogos em declínio")

sns.barplot(x = df["Ano"], y = df["Vendas Globais"])
plt.ylabel('Vendas globais em milhões');
```
